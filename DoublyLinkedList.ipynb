```python
# Doubly Linked List

class node:
    def __init__(self,data = None):
        self.data = data
        self.left = None
        self.right = None

class linked_Lists:
    def __init__(self):
        self.head_Node = node()
        
    def append(self,data):
        new_Node = node(data)
        cur_Node = self.head_Node
        while(cur_Node.right !=None):
            cur_Node = cur_Node.right
        cur_Node.right = new_Node
    
    def display(self):
        elements = []
        cur_Node = self.head_Node
        while(cur_Node.right !=None):
            cur_Node = cur_Node.right
            elements.append(cur_Node.data)
        return(elements)
    
    def insert(self,data,index):
        new_Node = node(data)
        if(index == 0):
            head = self.head_Node
            new_Node.right = head.right
            head.right.left = new_Node
            head.right = new_Node
            return
        cur_idx = 0
        new_Node = node(data)
        cur_Node = self.head_Node
        while(True):
            cur_Node = cur_Node.right
            if(cur_idx == index-1):
                new_Node.left = cur_Node
                new_Node.right = cur_Node.right
                cur_Node.right.left = new_Node
                cur_Node.right = new_Node
                return
            cur_idx += 1
     
    def length(self):
            count = 0
            cur_Node = self.head_Node
            while(cur_Node.right !=None):
                cur_Node = cur_Node.right
                count += 1
            return(count)
    
    def reverse(self):
        elements = self.display()
        cur_Node = self.head_Node
        cur_Node = cur_Node.right
        for i in range(len(elements) - 1,-1,-1):
            cur_Node.data = elements[i]
            cur_Node = cur_Node.right
        return
            
            
        
    
    def delete(self,index):
        
        if(index == self.length() - 1):
            cur_idx = 0
            cur_Node = self.head_Node
            while(True):
                last_Node = cur_Node
                cur_Node = cur_Node.right
                if(cur_idx == index):
                    last_Node.right = None
                    cur_Node.left = None
                    return
                cur_idx += 1
            
        else:
            cur_idx = 0
            cur_Node = self.head_Node
            while(True):
                last_Node = cur_Node
                cur_Node = cur_Node.right
                if(cur_idx == index):
                    last_Node.right = cur_Node.right
                    cur_Node.right.left = last_Node
                    return
                cur_idx += 1
            
    
linkedLists = linked_Lists()
linkedLists.append(1)
linkedLists.append(2)
linkedLists.append(3)
print(linkedLists.display())
linkedLists.insert(5,0)
linkedLists.insert(6,1)
print(linkedLists.display())
#linkedLists.delete(3)
print(linkedLists.display())
linkedLists.reverse()
print(linkedLists.display())



    
```
